```python
import numpy as np

mat = np.random.randint(1,100, size=(100,))
arr = np.where(mat>20)

print(mat)

value_grater_than_20 = np.count_nonzero(mat>20)
print(value_grater_than_20)
pp = np.unique(arr)
print(len(pp))
```
