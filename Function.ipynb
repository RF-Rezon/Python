```python
def squareit(a,b):
    return a**b
```

```python
## Positional Argument
x = squareit(10,2);
print(x)
```

```python
## Keyword Argument
x = squareit(b=3,a=10);   ## Keyword Argument er khetre age piche argument pathaleo prob nai. 
print(x)
```

```python
def squareit2(*args):   # args holo onekta rest operator er moto. 
    for i in args:
        print(i)
    
x = squareit2(10, 30, 40, 50);   ## Ekta tupple return kore. So iterate kore dekhano jabe.  
print(x)
```

```python
def squareit2(**args):  
    for key,value in args.items():   ## Dictionary method items use korle list akare back dey. Takhon iterate kora possible hy.
        print(key, value)            ## Dictionary direct iterate kora jay na. 
    
x = squareit2(name = "Rezon", roll= 10, section= "a")  
print(x)
```

## Return Types

- String

```python
## Single string 
def func():
    return "this is a string"

print(type(func()))
func()
```

```python
## Multiple strings 
def func():
    return "this is a string", "second string"

print(type(func()))

a,b = func()   ## Unpack korlam
print(a,b)
```

- List

```python
## Single list
def func():
    return [10, "Some text", 10.40, {40,30,20} ]

print(type(func()))
x = func()
for item in x:
    print(item)

print("------------------------------------") 

a,b,c,d = func()
print(a)
print(b)
print(c)
print(d)                 ## Tar mane Unpacking Sting, List er khetreo kaj kore. 
```

```python
## Multiple lists
def func():
    return [10, "Some text", 10.40, {40,30,20} ], [100, "Some text000", 100.400, {400,300,200} ]

print(type(func()))
x = func()
for item in x:
    print(*item)
```
