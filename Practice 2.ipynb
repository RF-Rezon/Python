```python
a_friends = {"Sakib", "Rahim", "Karim", "Jamal"}
b_friends = {"Sakib", "Jamal", "Rafiq", "Nadim"}

x = a_friends.intersection(b_friends)
print(x)
a_friends.difference_update(x)
print(a_friends)
b_friends.difference_update(x)
print(b_friends)
print(f"Total unique friends: {len(a_friends) + len(b_friends) + len(x)}")
```

```python
students = [("Rafi", 89), ("Sumi", 95), ("Hasan", 90), ("Nila", 75), ("Anik", 98)]

l = [score for name, score in students]
l.sort(reverse=True)

top_3 = l[:3]

for name, marks in students:
    if marks in top_3:
        print(f"{name} - {marks}")
```
