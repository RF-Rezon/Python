```python
## ---------------------- Designer -----------------------
class Laptop:
    def __init__(self, name, color, price, mac):
        self.name = name
        self.color = color
        self.__price = price
        self.__mac = mac

    def seePrice(self):
        return self.__price
        
    def seeMac(self):
        return self.__mac
        
    def changeMac(self, val):
        self.__mac = val

    
lenovo = Laptop("Lenevo", "black", "20k", "0x16")

## ---------------------- User -----------------------

print(lenovo.seeMac())
lenovo.changeMac("0xxxxxx10")
print(lenovo.seeMac())
```
