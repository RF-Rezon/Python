```python
# ----- Level 1 -----
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):
        print("Brand:", self.brand)


# ----- Level 2 -----
class Car(Vehicle):
    def __init__(self, brand, model):
        Vehicle.__init__(self, brand)   # <-- super() ছাড়া parent call
        self.model = model

    def show_model(self):
        print("Model:", self.model)


# ----- Level 3 -----
class SportsCar(Car):
    def __init__(self, brand, model, top_speed):
        Car.__init__(self, brand, model)   # <-- parent class নাম ধরে কল
        self.top_speed = top_speed

    def show_speed(self):
        print("Top Speed:", self.top_speed, "km/h")



# ----- Object -----
sc = SportsCar("Audi", "R8", 320)

sc.show_brand()   # Level 1
sc.show_model()   # Level 2
sc.show_speed()   # Level 3
```

```python
# ----- Level 1 (Grandparent Class) -----
class Vehicle:
    def __init__(self, brand):
        self.brand = brand               # Level 1 variable

    def show_brand(self):                # Level 1 method
        print(f"Brand: {self.brand}")


# ----- Level 2 (Parent Class) -----
class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)          # inherit Level 1 variable
        self.model = model               # Level 2 variable

    def show_model(self):                # Level 2 method
        print(f"Model: {self.model}")


# ----- Level 3 (Child Class) -----
class SportsCar(Car):
    def __init__(self, brand, model, top_speed):
        super().__init__(brand, model)   # inherit Level 2 + Level 1
        self.top_speed = top_speed       # Level 3 variable

    def show_speed(self):                # Level 3 method
        print(f"Top Speed: {self.top_speed} km/h")



# ----- Create object of Level 3 (Child) -----
sc = SportsCar("BMW", "M4", 280)

# Calling methods from all levels
sc.show_brand()   # from Level 1
sc.show_model()   # from Level 2
sc.show_speed()   # from Level 3
```
