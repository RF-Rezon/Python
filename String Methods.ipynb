```python
x = "This is my Pen, Pen, PEN, pen pen pen pin pin pin"
# Lower Kora
y = x.lower()
# Upper kora
# y = x.upper()

# Check kora string er moddhe ase kina
print("pen" in y)
# Length of a string
print(len(x))
# Ekta substring koto step e pawa jay (Ekadhik thakle 1st er ta)
print(y.find("is"))
# Ekta substring koto step e pawa jay (Ekadhik thakle 2nd ta)
print(y.rfind("is"))
# Ekta substring kotobar ache seta count korbo
print(y.count("pin"))
# Replace kora
print(y.replace("pen", "Book"))
print(x) # Sting by default kintu immutable. Change obosthay show kora jay but nij jaygay change hy na. 
```
