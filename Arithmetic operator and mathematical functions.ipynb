```python
import numpy as np 
```

```python
# faster for vectorization

x = np.array([ 10 , 8 , 30 , 100 ])
y = np.array([ 2 , 3 ,4  ,5 ]) 

add = np.add(x,y) 
print(add)

sub = x-y 
print(sub)

mul = x*y 
print(mul) 

div = x/y 
print(div) 

remainder = x % y

print(remainder) 
```

```python
# mathematical function 

# trigonometry 

# sin , cos , tan takes values in radian. Era radian chara ar kichu bujhe na.  

angle_deg = 90
angle_rad = np.deg2rad(angle_deg)

# cos_val = np.cos(x)
# print(cos_val)  # Radient e debe

sin_val = np.sin(angle_rad)  
print(sin_val)
```

```python
#log 

x = np.array([ 10 , 8 , 16 , 100 ])
base_10_log_val = np.log10(x)
print(base_10_log_val)


two_diye_bhag = np.log2(x) 
print(two_diye_bhag)

sqrt= np.sqrt(x)
print(sqrt)
```

```python
# sum 
s = np.sum(x)
print(s)

#prefix sum 
cumul =np.cumulative_sum(x)
print(cumul)
```
