```python
import numpy as np

# 1st array
arr1 = np.array([1,2,3,4,5])

# 2nd array
arr2 = np.array([
    [1,2,3,4,5],
    [10,50,60,90,80]])

# 3rd array
arr3 = np.array([
    [[1,2,3,4,5],[10,40,50,90,20]],
         [[1,2,3,4,5],[10,50,60,90,80]],
    [[1,2,3,4,5],[10,50,60,90,80]]
                ])

print (arr1.ndim)
print (arr2.ndim)
print (arr3.ndim)

print (arr1.shape)
print (arr2.shape)
print (arr3.shape)

print (arr1.dtype)
print (arr2.dtype)
print (arr3.dtype)

print (arr1.size)
print (arr2.size)
print (arr3.size)
```
