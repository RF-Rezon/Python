```python
# ----- Common Base Class -----
class Vehicle:
    def __init__(self, name="Generic Vehicle"):
        self.name = name

    def show_info(self):
        print(f"This is a {self.name}")


# ----- Derived Classes -----
class Car(Vehicle):
    def __init__(self, name="Car"):
        super().__init__(name)   # Base class constructor call
        # Car specific init code লাগলে এখানে লিখা যাবে


class SportsCar(Vehicle):
    def __init__(self, name="Sports Car"):
        super().__init__(name)
        self.top_speed = 300     # Extra attribute

    # Override
    def show_info(self):
        print(f"{self.name} has top speed {self.top_speed} km/h")


class Truck(Vehicle):
    def __init__(self, name="Truck"):
        super().__init__(name)
        self.capacity = "10 tons"

    # Override
    def show_info(self):
        print(f"{self.name} can carry {self.capacity}")


# ----- Objects -----
v = Vehicle()
c = Car()
sc = SportsCar()
t = Truck()

# ----- Polymorphic Calls -----
for vehicle in (v, c, sc, t):
    vehicle.show_info()
```
