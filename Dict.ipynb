- Key hishebe jekono immmutable like (int, string, float, tuple) use kora jay. Ar value hisebe immutable-mutable sob.
- So, Key gula immmutable but valugula mutable.

```python
student = {"name":"Rejwan", "age":25, (10): "handsome", "info": ["Nature", 20.14, True, "121313"]}
print(student, type(student))
```

## Access

```python
print(student["name"])
print(student["info"])

# using get method
print(student.get(10))
```

```python
# Change Value

student[10] = "Gym"     ## Mane immutable key ke dhore mutable value ke icchamoto change kora jay!
student["info"][2] = [True, False]

## Adding key value
student["new_key"] = {"name":"Raihan Dadu", "Age": 90}

print(student)
```

## Methods

```python
## Removes the element with the specified key
student.pop("name")
print(student)
```

## View Methods

```python
student = {"name":"Rejwan", "age":25, (10): "handsome", "info": ["Nature", 20.14, True, "121313"]}
# print(student.keys())
# print(student.values())
_list = student.items()
print(_list)
for key, value in _list:
    print(key, value)               ## Mane dictionary ke list baniye evabe iterate kora jay!
```

## Dict Comprehension

```python
make_dict = {x:x**2 for x in range(1,11)}
print(make_dict)
```

```python
co_ordinates = [(10,10.5),(20.5, 192),(101, 102)]
location = ["dhaka", "chottogram", "sylhet"]

exact_locaion = {lo:co for lo,co in zip(location,co_ordinates)}
print(exact_locaion)

## Evabe 2 ta loop eksathe dorkar hole bind korte "zip" use korte hobe.
```
