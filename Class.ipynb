## Class Attribute

```python
class Car:
    category = "Sedan"

audi = Car() 
marsedis = Car() 
bmw = Car() 
print(audi.category)
print(marsedis.category)
print(bmw.category)
```

```python
class Car:
    category = "Sedan"

audi = Car() 
marsedis = Car() 
bmw = Car()
bmw.category = "Racing"   #Changing
print(audi.category)
print(marsedis.category)
print(bmw.category)  # Change hoye gese
```

## Instance Attribute

```python
class Car:
    category = "Sedan"

    # Class er moddhei ekta constructor function banabo
    def __init__(i, model, seat, power):
        i.model = model
        i.seat = seat
        i.power = power  # self, i ja khusi nam deya jay


audi = Car("mx-5", 4, 200) 
marsedis = Car("db-12", 2, 300) 
bmw = Car("i-8", 4, 260) 

print(audi.power)
print(bmw.power)
```

```python
class Car:
    category = "Sedan"

    # Class er moddhei ekta constructor function banabo
    def __init__(i, model, seat, power, charge=100):
        i.model = model
        i.seat = seat
        i.power = power  # self, i ja khusi nam deya jay
        i.charge = charge

    def run (i, km):
        if(i.charge <= 0):
            print("No Charge.")
            return
        i.charge -= km * 2
        print(f"Bormonate tel ase {i.charge}") 
        
audi = Car("mx-5", 4, 200) 
marsedis = Car("db-12", 2, 300) 
bmw = Car("i-8", 4, 260) 

audi.run(2)
# audi.charge
marsedis.charge
```
