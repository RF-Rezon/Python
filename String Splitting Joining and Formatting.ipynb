### Splitting

```python
person1 = "My Name is Rafiq"
person2 = "My-Name-is-Jabbar"
x = person1.split()
y = person2.split("-")
print(x)
print(y)
```

### Joining

```python
a = "$".join(x)
print(a)
b = "-----".join(y)
print(b)
```
