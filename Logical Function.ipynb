## Logical Function

```python
import numpy as np 

# comparison - >,< , >= ,<= , == , !=

x = np.array([ 10 , 8 , 16 , 100 ])
y = np.array([ 2 , 3 ,16  ,5 ]) 

greater_than = x>y 

print(greater_than)

equal = x==y 

print(equal) 

# all () -> shob true kina 
# any() -> jekono ekta true ase kina 

print(np.all(greater_than)) 

print(np.any(equal))
```
