```python
import numpy as np 

arr = np.random.randint(1 ,100 , size =(10,5)) # sample - 10 , feature - 5

print(arr)
```

```python
b = arr.reshape(5,10)

print(b)
```

```python
flatten = b.flatten()

print(flatten)
```

```python
column_wise_flattening = np.ravel(b,order= 'C') 

print(column_wise_flattening) 
```

```python
column_wise_flattening = np.ravel(b,order= 'F') 

print(column_wise_flattening) 
```

```python
a = np.random.randint(1 ,10 , size =(2,3))
b = np.random.randint(20 ,30 , size =(2,3))
print(a)
print(b)
```

```python
con_row = np.concatenate((a,b),axis=0) # row wise 

print(con_row)
```

```python
con_row = np.concatenate((a,b),axis=1) # row wise 

print(con_row)
```

```python
# error 

# row wise concetanate -> column same 
# column wise concetanate -> row same 

x = np.random.randint(1 ,10 , size =(2,3))
y = np.random.randint(1 ,10 , size =(1,3))

r = np.concatenate((x,y),axis =0 )
r = np.concatenate((x,y),axis =1 )
print(r)
```

```python
# transpose 

mat = np.array([[10,20,30],
                [30,40,50]
               ])


transpose = mat.T

print(transpose) 
```

```python
# array split 

a = np.random.randint(1 ,10 , size =(10,))
print(a)
splitted_array = np.array_split(a,3)  # soman soman vag korte na parle error debe na. 

print(splitted_array)  

# split -> equal division 
splitted_array = np.split(a,3)   # soman soman vag korte na parle error debe. 
print(splitted_array)
```
