## For loop ke short kore "map"
## Function ke short kore "lambda"

```python
## With Normal Function
```

```python
x = [1,2,3,4,5]

def func (arg):
    y= []
    for i in arg:
        y.append(i**2)
        
    return y

func(x)
```

```python
## With Map
```

```python
x = [1,2,3,4,5]

def func (arg):
    return arg**2
    
z = list(map(func, x))
print(z)
```

```python
## With Map and Lambda
```

```python
x = [1,2,3,4,5]

z = list(map(lambda x: x**2, x))
print(z)
```
