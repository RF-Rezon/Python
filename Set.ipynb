### Characteristics

- Unordered collection (set-এর কোনো নির্দিষ্ট ক্রম (order) থাকে না)
- Always duplicate free thake
- Mutable
- Indexing, slicing nai

### Methods

```python
x = {10,30,40,50,60,100}
x.add(300) # List e "append", set e "add". "append" korle sure last e. But "add" e surity nai. 
print(x)
x.pop() # List e pop last er ta pop kore. Set e pop first er ta. 
print(x)
```

### Mathematical Methods

```python
q = {10, 20}
p = {10, 20, 40, 50, 70}

print(p.union(q))
print(p.intersection(q))
print(q.issubset(p)) 
print(q.isdisjoint(p)) # 2 tar moddhe kono match na thakle true
```
