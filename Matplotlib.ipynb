```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)
y = x ** 2
```

```python
plt.plot(x,y)
plt.title("Chart")
plt.xlabel("Time")
plt.ylabel("Human Hair")
```

# Subplot

```python
plt.subplot(3,3,2)    # 3 ta row, 3 ta column, select korsi 2 number ghorta. 
plt.plot(x,y+3)
plt.subplot(3,3,1)    # 3 ta row, 3 ta column, select korsi 1 number ghorta. 
plt.subplot(3,3,5)    # 3 ta row, 3 ta column, select korsi 5 number ghorta. 
plt.plot(y, x)
plt.subplot(3,3,3)    # 3 ta row, 3 ta column, select korsi 3 number ghorta. 
```

# Object Oriented Way

```python
fig = plt.figure(dpi=600)
box1 = fig.add_axes([0.2, 0.2, 1, 1])
box1.plot(x,y)
box2 = fig.add_axes([0.2, 0.2, 0.6, 0.6])
box2.plot(x,y)
box3 = fig.add_axes([0.2, 0.2, 0.4, 0.4])
box3.plot(x,y)
```

```python
fig = plt.figure()
box1 = fig.add_axes([0.2, 0.2, 1, 1])
box1.plot(x,y)
box2 = fig.add_axes([0.5, 0.5, 0.6, 0.6])
box2.plot(x,y)
box3 = fig.add_axes([0.3, 0.5, 0.4, 0.4])
box3.plot(x,y)
```
