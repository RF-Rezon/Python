## List As Stack And Queue

## Stack

```python
stack = []
stack.append(1)
stack.append(2)
stack.append(3)
stack.append(4)
stack.append(5)
print(f"Full stack is: {stack}")
print(f"Last element is: {stack[-1]}")
stack.pop()
print(f"Last element is: {stack[-1]}")
stack.pop()
print(f"Last element is: {stack[-1]}")
```

## Queue

```python
queue = []

queue.append(1)
queue.append(2)
queue.append(3)
queue.append(4)
queue.append(5)
print(f"Full queue is: {queue}")
print(f"First element is: {queue[0]}")
queue.pop(0)
print(f"First element is: {queue[0]}")
queue.pop(0)
print(f"First element is: {queue[0]}")
```

### List Comprehension

```python
# >>>>>>>>>>>>>>>>>>>>>>  Naive System
list_one = []

for i in range(101):
    if i%2 != 0:
        list_one.append(i)
print(list_one)
# >>>>>>>>>>>>>>>>>>>>>>>>>>>>>>  List Comprehension System
random_list = [x for x in range(0,101) if x%2 != 0] 
print(random_list)
```
