```python
import numpy as np
```

```python
#diag

inf = np.diag([1,3,5,7,9])
print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)
```

```python
#identity 1

inf = np.eye(5)  # row-column same. 
print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)
```

```python
#identity 2

inf = np.eye(5,7)  # row-column not same. 
print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)
```

```python
#identity 2

inf = np.eye(5,7,3)  # row-column not same. 
print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)
```
