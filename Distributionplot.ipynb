```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = sns.load_dataset('tips')
df             # seaborn er github theke asche. 
```

# Distplot/Histplot

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

# jointplot

```python
sns.jointplot(x=df['total_bill'], y=df['tip'], kind='hex')
```

```python
sns.jointplot(x=df['total_bill'], y=df['tip'], kind='scatter')
```

```python
sns.pairplot(df)
```

```python
sns.rugplot(df)
```

```python
sns.rugplot(df['tip'])
```
