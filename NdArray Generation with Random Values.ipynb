```python
import numpy as np

inf = np.random.rand(2,3) # 0 -> 1 (float generate kore)   # (shape)

print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)
```

```python
inf = np.random.randint(1,10, (2,3))  # (range, (shape))

print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)
```

```python
inf = np.random.uniform(50,100, (2,3))  # (range, (shape))

print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)

inf = np.random.uniform(50,100, (3,3,3))  # (range, (shape))

print(inf)
print(type(inf))
print(inf.shape)
print(inf.size)
print(inf.ndim)
print(inf.dtype)
```
